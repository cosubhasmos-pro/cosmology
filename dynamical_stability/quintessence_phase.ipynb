## Plotting Phase Space for Model 

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Define the system. 
The system is obtained by taking variables say $x$ and $y$, and they transform the Friedmann equations and the scalar field equation into an autonomous system of first-order differential equations.

```python
def dx_dN(x, y, lambda_=1):
    return -3*x + 0.5*np.sqrt(6)*lambda_*y**2 + x*(4*x**2 - 2*y**2)

def dy_dN(x, y, lambda_=1):
    return -0.5*np.sqrt(6)*lambda_*x*y + y*(4*x**2 - 2*y**2)
```

### Creating Grid and Computing Derivative

```python
x = np.linspace(-1.5, 1.5, 20)
y = np.linspace(-1.5, 1.5, 20)
X, Y = np.meshgrid(x, y)

U = dx_dN(X, Y)
V = dy_dN(X, Y)
```

```python
plt.figure(figsize=(8, 8))
plt.streamplot(X, Y, U, V, density=1.5)
plt.plot(0, 0, 'ro', label='Matter-dominated (0,0)')
plt.plot(1, 0, 'bo', label='Kinetic-dominated (1,0)')
plt.plot(-1, 0, 'bo')
plt.plot(0.408, 0.913, 'go', label='Scalar field-dominated (0.408, 0.913)')
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Phase Portrait of Quintessence Model (=1)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.streamplot(X, Y, U, V, density=1.5)
plt.plot(0, 0, 'ro', label='Matter-dominated (0,0)')
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Phase Portrait of Quintessence Model (=1)')
plt.legend()
plt.grid(True)
plt.show()
```
